```python
import numpy as np
import matplotlib.pyplot as plt
Temp = np.linspace(0.001, 1, 1000)
sigma0=1.0 
T0=1.0
pi=np.pi
sigma=sigma0*np.exp(-(T0/Temp)**0.25)
lnsigma=np.log(sigma)
Tvar=Temp**0.25
TvarN=Temp**(-0.25)
#Plot
plt.figure(figsize=(10,6))
plt.plot(sigma, Temp, 'r*',label = "Variable Hopping", linewidth=1.0)
#plt.yscale("log")
#plt.xscale("log")
plt.title("Sigma")
plt.xlabel("Temp")
plt.ylabel("Variable Hopping")
plt.grid(visible=None, which='both', axis='both')
plt.show()
#print (sigma)
```

```python
plt.figure(figsize=(10,6))
plt.plot(lnsigma, Tvar, 'r*',label = "Variable Hopping", linewidth=1.0)
#plt.yscale("log")
#plt.xscale("log")
plt.title("Sigma")
plt.xlabel("Temp^0.25")
plt.ylabel("Variable Hopping")
plt.grid(visible=None, which='both', axis='both')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(lnsigma, TvarN, 'r*',label = "Variable Hopping", linewidth=1.0)
#plt.yscale("log")
#plt.xscale("log")
plt.title("Sigma")
plt.xlabel("Temp^-0.25")
plt.ylabel("Variable Hopping")
plt.grid(visible=None, which='both', axis='both')
plt.show()
```
